```python
from urllib.request import urlopen
import json
import pandas as pd
```

```python
response = urlopen('https://api.openf1.org/v1/sessions')
data = json.loads(response.read().decode('utf-8'))
print(data)
```

```python
sessions = pd.DataFrame(data)
```

```python
sessions
```

```python
sessions.session_type.unique()
```

```python
# Delete rows with session type is Practice
sessions = sessions[sessions.session_type != 'Practice']
```

```python
sessions.session_key.unique()
```

```python
response= urlopen('https://api.openf1.org/v1/drivers')
data = json.loads(response.read().decode('utf-8'))
df = pd.DataFrame(data)
df
```

```python
df.session_key.unique()
```

```python
df = df[df.session_key.isin(sessions.session_key.unique().tolist())]
```

```python
drivers = df[['session_key', 'driver_number']]
drivers
```

```python
laps = pd.DataFrame(columns=['meeting_key', 
  'session_key', 
  'driver_number', 
  'i1_speed', 
  'i2_speed', 
  'st_speed', 
  'date_start', 
  'lap_duration', 
  'is_pit_out_lap', 
  'duration_sector_1', 
  'duration_sector_2', 
  'duration_sector_3', 
  'segments_sector_1', 
  'segments_sector_2', 
  'segments_sector_3', 
  'lap_number'])
```

```python
from urllib.error import HTTPError
import time

for session_key, driver_number in drivers[['session_key', 'driver_number']].values:
    print('Downloading data for driver', driver_number, 'in session', session_key)
    for lap in range(2, 80):
        for attempt in range(3):  # 3 tries
            try:
                response = urlopen('https://api.openf1.org/v1/laps?session_key=' + str(session_key) + '&driver_number=' + str(driver_number) + '&lap_number=' + str(lap))

                data = json.loads(response.read().decode('utf-8'))

                if len(data) == 0:
                    break
                
                lap_data = pd.json_normalize(data)
                laps = pd.concat([laps, lap_data], ignore_index=True)
                
                break 
            except HTTPError as e:
                if e.code == 429:
                    print("Rate limit reached. Waiting before retrying...")
                    time.sleep(5)
                else:
                    print(f"HTTP Error: {e.code}. Skipping lap {lap} for driver {driver_number}.")
                    break
        else:
            print(f"Failed to fetch lap {lap} for driver {driver_number} after 3 attempts.")
```

```python
laps
```

```python
laps.to_csv('laps.csv', index=False)
```
